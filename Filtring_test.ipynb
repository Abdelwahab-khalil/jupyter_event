```python
import pandas as pd
from scipy import sparse
```

```python
ratings = pd.read_csv('ratings.csv')
event = pd.read_csv('event.csv')
ratings = pd.merge(event,ratings)
ratings.head()
```

```python
ratings = pd.merge(event,ratings).drop(['event_date','event_city','event_time','event_cat'],axis=1)
ratings.head()
```

```python
userRatings = ratings.pivot_table(index=['user_id'],columns=['event_name'],values='rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix.head(100)
```

```python
def get_similar(event_name,rating):
    similar_ratings = corrMatrix[event_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    print(type(similar_ratings))
    return similar_ratings
```

```python

event_rating = [("The Royal Agricultural Winter Fair",5),("Big Data Toronto (BDT)",1),("Understanding Emotional Intelligence",5)]
similar_event = pd.DataFrame()
for event,rating in event_rating:
    similar_event = similar_event.append(get_similar(event,rating),ignore_index = True)

similar_event.head(10)
```
